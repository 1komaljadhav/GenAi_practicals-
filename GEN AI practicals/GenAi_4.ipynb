Build a simple question-answering chatbot using a pre-trained language model.

```python
from transformers import pipeline

# Load pre-trained Question-Answering model
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

def answer_question(context, question):
    """ Answers a question based on the given context using a pre-trained model. """
    response = qa_pipeline(question=question, context=context)
    return response["answer"]

# Sample Context (You can replace it with a real knowledge base)
context = """
ChatGPT is an AI model developed by OpenAI based on the GPT architecture.
It is used for conversational AI, question answering, and text generation.
The model has been fine-tuned using Reinforcement Learning from Human Feedback (RLHF).
"""

# Chatbot loop
print("Chatbot: Ask me anything based on context ! Type 'exit' to quit.")
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Chatbot: Goodbye!")
        break

    response = answer_question(context, user_input)
    print(f"Chatbot: {response}")
```
