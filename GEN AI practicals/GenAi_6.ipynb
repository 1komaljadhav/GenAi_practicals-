```python
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt

# Load the pre-trained Stable Diffusion model
model_id = "runwayml/stable-diffusion-v1-5"  # You can use different versions
pipeline = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipeline.to("cuda" if torch.cuda.is_available() else "cpu")  # Use GPU if available

# Function to generate and display images from text
def generate_image(prompt, num_images=1):
    images = pipeline(prompt, num_inference_steps=50, guidance_scale=7.5).images  # Generate image
    for i, img in enumerate(images[:num_images]):
        plt.figure(figsize=(6, 6))
        plt.imshow(img)
        plt.axis("off")
        plt.show()

# Example Usage
prompt = "A futuristic city with flying cars at sunset"
generate_image(prompt)
```
